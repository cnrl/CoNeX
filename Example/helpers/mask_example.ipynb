```python
import conex
import matplotlib.pyplot as plt
from torchvision import datasets
from torchvision.transforms import ToTensor, Compose
```

```python
m = 2
n = 2
gap = (-5,-5,-5,-5)
sample_index = 1
```

```python
grid_erase_transforms = Compose([ToTensor(), conex.GridEraseMask(m=m, n=n, gap=gap)])
grid_keep_transforms = Compose([ToTensor(), conex.GridKeepMask(m=m, n=n, gap=gap)])
grid_crop_transforms = Compose([ToTensor(), conex.GridCropMask(m=m, n=n, gap=gap)])
```

```python
mnist = datasets.MNIST(root="Dataset", transform=ToTensor(), download=True)
mnist_erase = datasets.MNIST(root="Dataset", transform=grid_erase_transforms, download=True)
mnist_keep = datasets.MNIST(root="Dataset", transform=grid_keep_transforms, download=True)
mnist_crop = datasets.MNIST(root="Dataset", transform=grid_crop_transforms, download=True)
```

```python
plt.imshow(mnist[sample_index][0][0])
plt.show()
```

```python
fig, axes = plt.subplots(m,n, figsize=(10,10))
for i in range(m):
    for j in range(n):
        axes[i][j].imshow(mnist_erase[sample_index][0][0][i*n+j][0])
plt.show()
```

```python
fig, axes = plt.subplots(m,n, figsize=(10,10))
for i in range(m):
    for j in range(n):
        axes[i][j].imshow(mnist_keep[sample_index][0][0][i*n+j][0])
plt.show()
```

```python
fig, axes = plt.subplots(m,n, figsize=(10,10))
for i in range(m):
    for j in range(n):
        axes[i][j].imshow(mnist_crop[sample_index][0][0][i*n+j][0])
plt.show()
```
